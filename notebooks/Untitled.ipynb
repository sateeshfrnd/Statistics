```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gamma, expon
```

```python
# Generate random data from normal distribution
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Generate random data from gamma distribution
data_gamma = np.random.gamma(shape=2, scale=1, size=1000)

# Generate random data from exponential distribution
data_exponential = np.random.exponential(scale=1, size=1000)
```

```python
plt.figure(figsize=(10, 5))

# Plot histograms
# plt.hist(data_normal, bins=30, density=True, alpha=0.5, color='blue', label='Normal Distribution (Histogram)')
# plt.hist(data_gamma, bins=30, density=True, alpha=0.5, color='green', label='Gamma Distribution (Histogram)')
# plt.hist(data_exponential, bins=30, density=True, alpha=0.5, color='orange', label='Exponential Distribution (Histogram)')

# Plot line graphs
x = np.linspace(-3, 8, 1000)
plt.plot(x, norm.pdf(x, loc=0, scale=1), linestyle='-', color='blue', label='Normal Distribution (PDF)')
plt.plot(x, gamma.pdf(x, a=2, loc=0, scale=1), linestyle='-', color='green', label='Gamma Distribution (PDF)')
plt.plot(x, expon.pdf(x, loc=0, scale=1), linestyle='-', color='orange', label='Exponential Distribution (PDF)')

plt.title('Distributions Comparison')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gamma

# Generate random data from normal distribution
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Generate random data from gamma distribution
data_gamma = np.random.gamma(shape=2, scale=1, size=1000)

# Plot histograms to visualize distributions
plt.figure(figsize=(10, 5))

# Plot histogram for normal distribution
plt.hist(data_normal, bins=30, density=True, alpha=0.5, color='blue', label='Normal Distribution (Mesokurtic)')

# Plot histogram for gamma distribution
plt.hist(data_gamma, bins=30, density=True, alpha=0.5, color='green', label='Gamma Distribution (Leptokurtic)')

# Plot normal distribution PDF for comparison
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, loc=0, scale=1), linestyle='-', color='blue', label='Normal Distribution PDF')

# Plot gamma distribution PDF for comparison
x = np.linspace(0, 10, 1000)
plt.plot(x, gamma.pdf(x, a=2, loc=0, scale=1), linestyle='-', color='green', label='Gamma Distribution PDF')

plt.title('Distributions with Low and High Kurtosis')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, gamma, expon

# Generate random data from normal distribution
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Generate random data from gamma distribution
data_gamma = np.random.gamma(shape=2, scale=1, size=1000)

# Generate random data from exponential distribution
data_exponential = np.random.exponential(scale=1, size=1000)

# Plot histograms to visualize distributions
plt.figure(figsize=(10, 5))

# Plot histogram for normal distribution
plt.hist(data_normal, bins=30, density=True, alpha=0.5, color='blue', label='Normal Distribution (Mesokurtic)')

# Plot histogram for gamma distribution
plt.hist(data_gamma, bins=30, density=True, alpha=0.5, color='green', label='Gamma Distribution (Leptokurtic)')

# Plot histogram for exponential distribution
plt.hist(data_exponential, bins=30, density=True, alpha=0.5, color='orange', label='Exponential Distribution (Platykurtic)')

# Plot normal distribution PDF for comparison
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, loc=0, scale=1), linestyle='--', color='blue', label='Normal Distribution PDF')

# Plot gamma distribution PDF for comparison
x = np.linspace(0, 10, 1000)
plt.plot(x, gamma.pdf(x, a=2, loc=0, scale=1), linestyle='--', color='green', label='Gamma Distribution PDF')

# Plot exponential distribution PDF for comparison
x = np.linspace(0, 5, 1000)
plt.plot(x, expon.pdf(x, loc=0, scale=1), linestyle='--', color='orange', label='Exponential Distribution PDF')

plt.title('Distributions with Different Kurtosis')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
