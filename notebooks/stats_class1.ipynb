```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import warnings
```

## Normal Distribution/ Gaussion Distribution

```python
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution
```

```python
size = 1000  # Size of the dataset
data = np.random.normal(mean, std_dev, size)
```

```python
plt.hist(data, 20, density = True)
plt.show()
```

```python
sns.distplot(data)
plt.show()
```

### Uniform distribution 
All the outcomes has equal probability to occur in range

```python
uniform_dist_data = np.random.uniform(1, 0, size)
```

```python
plt.hist(uniform_dist_data)
```

### Log Normal Distribution

Whose Logarithum is normally distributed.

```python
log_noraml_dist_data = np.random.lognormal(mean = 0, sigma = 0.5, size = 1000)
```

```python
plt.hist(log_noraml_dist_data)
```

```python
log_noraml_dist_data1 = log_noraml_dist_data[np.where(abs(log_noraml_dist_data) < 5)]
```

```python
plt.hist(log_noraml_dist_data, bins=100)
plt.show()
```

```python
sns.distplot(log_noraml_dist_data1)
plt.show()
```

Right Skew data

#### Convert log normal distribution to Normal Distribution

```python
sns.distplot(np.log(log_noraml_dist_data1))
plt.show()
```

#### Square Root Trnasformation 

```python
sns.distplot(np.sqrt(log_noraml_dist_data1))
plt.show()
```

## Chisquare Distribution

```python
chis_dist = np.random.chisquare(df = 1, size = 10000)
chis_dist = chis_dist[np.where(abs(chis_dist) < 10)]

# plt.distplot(chis_dist, bins = 1000)
sns.distplot(chis_dist)
plt.show()
```

```python
chis_dist = np.random.chisquare(df = 2, size = 10000)
chis_dist = chis_dist[np.where(abs(chis_dist) < 10)]

# plt.hist(chis_dist, bins = 1000)
sns.distplot(chis_dist)
plt.show()
```

```python
chis_dist = np.random.chisquare(df = 5, size = 10000)
chis_dist = chis_dist[np.where(abs(chis_dist) < 10)]

# plt.hist(chis_dist, bins = 1000)
sns.distplot(chis_dist)
plt.show()
```

### T Distribution

```python
t_dist = np.random.standard_t(df = 1, size = 10000)
t_dist = t_dist[np.where(abs(t_dist) < 10)]

# plt.hist(chis_dist, bins = 1000)
sns.distplot(t_dist)
plt.show()
```

```python
t_dist = np.random.standard_t(df = 2, size = 10000)
t_dist = t_dist[np.where(abs(t_dist) < 10)]

# plt.hist(chis_dist, bins = 1000)
sns.distplot(t_dist)
plt.show()
```

```python
t_dist = np.random.standard_t(df = 10, size = 10000)
t_dist = t_dist[np.where(abs(t_dist) < 10)]

# plt.hist(chis_dist, bins = 1000)
sns.distplot(t_dist)
plt.show()
```

### F-Distribution

```python
# dfnum is df for numerator, dfden is df for den
f_dist = np.random.f(dfnum = 1, dfden=1,  size = 10000)
f_dist = f_dist[np.where(abs(f_dist) < 10)]

sns.distplot(f_dist)
plt.show()
```

```python
# dfnum is df for numerator, dfden is df for den
f_dist = np.random.f(dfnum = 1, dfden=10,  size = 10000)
f_dist = f_dist[np.where(abs(f_dist) < 10)]

sns.distplot(f_dist)
plt.show()
```
